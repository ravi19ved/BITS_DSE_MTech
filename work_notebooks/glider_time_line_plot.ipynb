```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Previous glider deployment data
data = {
    'Glider': ['890', '890', '891', '891', '1096', '1097', '1126', '1127', '1127', '1128', '1129', '1130', '1131'],
    'Deployment_Date': ['2024-01-01', '2024-06-01', '2024-01-15', '2024-07-01', '2024-02-01', '2024-02-15', '2024-03-01', '2024-03-15', '2024-08-01', '2024-04-01', '2024-04-15', '2024-05-01', '2024-05-15'],
    'Retrieval_Date':   ['2024-01-31', '2024-06-30', '2024-02-14', '2024-07-31', '2024-02-28', '2024-03-14', '2024-03-31', '2024-04-14', '2024-08-30', '2024-04-30', '2024-05-14', '2024-05-31', '2024-06-14']
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert Deployment_Date and Retrieval_Date to datetime objects
df['Deployment_Date'] = pd.to_datetime(df['Deployment_Date'])
df['Retrieval_Date'] = pd.to_datetime(df['Retrieval_Date'])

# Calculate the operational days
df['Operational_Days'] = (df['Retrieval_Date'] - df['Deployment_Date']).dt.days

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

# Define a color for all bars
bar_color = '#1f77b4'  # A blue shade

# Create horizontal bars for each deployment
for idx, row in df.iterrows():
    ax.barh(row['Glider'], row['Operational_Days'], left=row['Deployment_Date'], height=0.4, color=bar_color)
    # Add the operational days as text slightly above the bars
    ax.text(row['Deployment_Date'] + pd.Timedelta(days=row['Operational_Days']/2), row['Glider'] + 0.4, f'{row["Operational_Days"]} days',
            va='center', ha='center', color='white', fontsize=9, fontweight='bold')
    # Add small vertical line for deployment date
    ax.vlines(row['Deployment_Date'], row['Glider'] - 0.2, row['Glider'] + 0.2, color='green', linewidth=2)
    # Add label for deployment date slightly above the bar
    ax.text(row['Deployment_Date'], row['Glider'] + 0.3, f'{row["Deployment_Date"].strftime("%Y-%m-%d")}', 
            va='center', ha='left', color='green', fontsize=8)
    # Add small vertical line for retrieval date
    ax.vlines(row['Retrieval_Date'], row['Glider'] - 0.2, row['Glider'] + 0.2, color='red', linewidth=2)
    # Add label for retrieval date slightly above the bar
    ax.text(row['Retrieval_Date'], row['Glider'] + 0.3, f'{row["Retrieval_Date"].strftime("%Y-%m-%d")}', 
            va='center', ha='left', color='red', fontsize=8)

# Adding grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Formatting date on x-axis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)

# Adding labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Gliders')
ax.set_title('Glider Operational Days with Multiple Deployments (Specific Dates)')

plt.tight_layout()
plt.show()
```
